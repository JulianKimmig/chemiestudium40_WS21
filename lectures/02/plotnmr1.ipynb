```python
import os # working with the operating system (e.g. files and folders)
import numpy as np # numpy, do all the math 
import matplotlib.pyplot as plt # plotting
```

```python
file = "../../data/spectrum_processed.csv" # realtive path to the data file from the notebook folder (".." for parent foler)
os.path.exists(file) # check if the file es there
```

```python
with open(file,"r") as f: # open the file in read mode
    content = f.read() # reads the content as a single string
content # display the content
```

```python
lines = content.split("\n") # splits the content by line linebreak
lines # show the linses
```

```python
header = lines[0].split(";") # the fistt line is the header, seperated by ";"
header
```

```python
data=[] # all the data
for line in lines[1:]: # iterate over all lines, except header
    linedata=[] # the data of the line
    if len(line)==0: # if the line is empty
        continue # jump to the next iteration in the for loop
    for value in line.split(";"): # for each value in the line, seperated by ;
        fval = float(value.replace(",",".")) # the value das a "," as a decimal, so we need to replace it with an ".", an the we read the string value as a flaot
        linedata.append(fval) # add the value to the linedata lsit
    data.append(linedata) # add the linedata list to the all data list
data # dispaly
```

```python
npdata = np.array(data) # transfer it to an array
npdata # dispaly
```

```python
#üplotting
plt.plot(
    npdata[:,0], #x values (":" all values of the first dimension, "0": the fist value of the second dimension)
    npdata[:,1], #x values (":" all values of the first dimension, "1": the second value of the second dimension)
    label="raw nmr" # label for the legend
)
plt.xlim(0,6) # the ploting limit of the x values ("zoom to range")
plt.xlabel(header[0]) # x-axis label
plt.ylabel(header[1]) # y-axis label
plt.title("sample nmr") # title of the plot
plt.legend() # create legend
```
